```python
import os
import glob
import numpy as np
import pandas as pd
import mdtraj as md
from pathlib import Path

# ── настройки ──────────────────────────────────────────────────────────────
BIOEMU_DIR      = "bioemu_output"
SCAFFOLDING_DIR = "inference"
METADATA_TSV    = "inpainting_dataset/metadata.tsv"
RMSD_THRESHOLD  = 3.5   # Ångström — порог "близко к конформации"
# ───────────────────────────────────────────────────────────────────────────


def load_pdb_ca(path):
    traj   = md.load(path)
    top    = traj.topology
    ca_idx = top.select("name CA")
    coords = traj.xyz[0, ca_idx, :]
    resids = np.array([top.atom(i).residue.resSeq for i in ca_idx])
    return coords, resids


def load_xtc_ca(xtc_path, top_path):
    traj   = md.load(xtc_path, top=top_path)
    ca_idx = traj.topology.select("name CA")
    return traj.xyz[:, ca_idx, :]


def rmsd_kabsch(ref, mob):
    ref = ref - ref.mean(0)
    mob = mob - mob.mean(0)
    H   = mob.T @ ref
    U, _, Vt = np.linalg.svd(H)
    d   = np.linalg.det(Vt.T @ U.T)
    D   = np.diag([1, 1, d])
    R   = Vt.T @ D @ U.T
    diff = (mob @ R.T) - ref
    return float(np.sqrt((diff**2).sum(-1).mean()))


def get_scaffold_mask(all_resids, motif_resids_set):
    return np.array([r not in motif_resids_set for r in all_resids])


def find_paths(sample_id):
    pattern = os.path.join(SCAFFOLDING_DIR,
                           f"inpainting_{sample_id}_{sample_id}",
                           f"job_0_id_0_motif_{sample_id}",
                           "*.pdb")
    pdbs   = sorted(glob.glob(pattern))
    path_a = next((p for p in pdbs if "pathA" in p), None)
    path_b = next((p for p in pdbs if "pathB" in p), None)
    return path_a, path_b


def find_bioemu(sample_id):
    # пробуем напрямую по sample_id, потом по вхождению в имя папки
    for cand in [str(sample_id)] + [d for d in os.listdir(BIOEMU_DIR)
                                     if str(sample_id) in d]:
        folder = os.path.join(BIOEMU_DIR, cand)
        if not os.path.isdir(folder):
            continue
        xtcs = glob.glob(os.path.join(folder, "*.xtc"))
        tops = glob.glob(os.path.join(folder, "*.pdb"))
        if xtcs and tops:
            return xtcs[0], tops[0]
    return None, None


def classify_frame(rmsd_a, rmsd_b, thr):
    """Куда ближе конкретный фрейм, с учётом порога."""
    near_a = rmsd_a < thr
    near_b = rmsd_b < thr
    if near_a and near_b:
        return "both"
    elif near_a:
        return "A"
    elif near_b:
        return "B"
    else:
        return "neither"


# ── основная логика ─────────────────────────────────────────────────────────

meta    = pd.read_csv(METADATA_TSV, sep="\t")
records = []

for _, row in meta.iterrows():
    sample_id    = row["sample_id"]
    gap_resseq   = set(map(int, str(row["gap_resseq"]).split(",")))

    path_a, path_b = find_paths(sample_id)
    if path_a is None or path_b is None:
        print(f"[{sample_id}] pathA/pathB не найдены — пропуск")
        continue

    xtc, top = find_bioemu(sample_id)
    if xtc is None:
        print(f"[{sample_id}] BioEmu output не найден — пропуск")
        continue

    ca_a, resids_a = load_pdb_ca(path_a)
    ca_b, resids_b = load_pdb_ca(path_b)
    frames         = load_xtc_ca(xtc, top)    # [4, n_res, 3]

    sc_mask = get_scaffold_mask(resids_a, gap_resseq)

    if frames.shape[1] != ca_a.shape[0]:
        print(f"[{sample_id}] Размерность не совпадает: "
              f"BioEmu={frames.shape[1]}, pathA={ca_a.shape[0]} — пропуск")
        continue

    sc_a = ca_a[sc_mask]
    sc_b = ca_b[sc_mask]

    rmsd_to_a, rmsd_to_b, labels = [], [], []
    for frame in frames:
        ra = rmsd_kabsch(sc_a, frame[sc_mask]) * 10   # → Å
        rb = rmsd_kabsch(sc_b, frame[sc_mask]) * 10
        rmsd_to_a.append(ra)
        rmsd_to_b.append(rb)
        labels.append(classify_frame(ra, rb, RMSD_THRESHOLD))

    rmsd_to_a = np.array(rmsd_to_a)
    rmsd_to_b = np.array(rmsd_to_b)

    self_rmsd   = rmsd_kabsch(sc_a, sc_b) * 10
    n_A         = labels.count("A")
    n_B         = labels.count("B")
    n_both      = labels.count("both")
    n_neither   = labels.count("neither")

    # ключевой критерий: BioEmu посетил оба состояния
    visits_both = (min(rmsd_to_a) < RMSD_THRESHOLD) and \
                  (min(rmsd_to_b) < RMSD_THRESHOLD)

    records.append({
        "sample_id"       : sample_id,
        "pathAB_rmsd"     : round(self_rmsd, 2),
        "min_rmsd_to_A"   : round(rmsd_to_a.min(), 2),
        "min_rmsd_to_B"   : round(rmsd_to_b.min(), 2),
        "mean_rmsd_to_A"  : round(rmsd_to_a.mean(), 2),
        "mean_rmsd_to_B"  : round(rmsd_to_b.mean(), 2),
        "frames_near_A"   : n_A,
        "frames_near_B"   : n_B,
        "frames_near_both": n_both,
        "frames_neither"  : n_neither,
        "frame_labels"    : labels,        # детальная картина
        "visits_both"     : visits_both,
    })

    print(f"[{sample_id}] AB_RMSD={self_rmsd:.1f}Å | "
          f"min→A={rmsd_to_a.min():.1f} min→B={rmsd_to_b.min():.1f} | "
          f"labels={labels} | visits_both={visits_both}")

# ── итоги ───────────────────────────────────────────────────────────────────
df = pd.DataFrame(records)

print("\n=== SUMMARY ===")
cols = ["sample_id","pathAB_rmsd","min_rmsd_to_A","min_rmsd_to_B",
        "frames_near_A","frames_near_B","frames_neither","visits_both"]
print(df[cols].to_string(index=False))

n_visits = df["visits_both"].sum()
print(f"\nПорог RMSD: {RMSD_THRESHOLD}Å")
print(f"Кандидатов, где BioEmu посетил оба состояния: {n_visits}/{len(df)}")
print(f"Среднее pathA-pathB RMSD: {df['pathAB_rmsd'].mean():.2f}Å")

# топ: большое AB расстояние + оба состояния посещены
top = df[df["visits_both"]].sort_values("pathAB_rmsd", ascending=False)
print(f"\nТоп кандидаты:")
print(top[cols].to_string(index=False))
```
